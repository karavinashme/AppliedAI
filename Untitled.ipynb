```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
hman=pd.read_csv('haberman.csv')
```

## 2D scatter

```python
hman.plot(kind='scatter', x='age', y='year');
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(hman,hue='status',size=4).map(plt.scatter, "age", "year").add_legend()
```

```python
plt.close()
```

```python
sns.relplot(x='age', y='year',data=hman, hue='status').add_legend()
```

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(hman,hue="status")
plt.show()
```
